```python
import torch
import esm_one_hot
from esm_one_hot import pretrained

%load_ext autoreload
%autoreload 2
```

```python
msa_tran, msa_tran_alphabet = pretrained.esm_msa1b_t12_100M_UR50S()
msa_tran_alphabet.all_toks
```

```python
esm2, esm2_alphabet = pretrained.esm2_t30_150M_UR50D()
esm2_alphabet.all_toks
```

```python
test_protein = [
    ('test1', 'MSDF'), ('test2', 'MPAF')
]

batch_converter = esm2_alphabet.get_batch_converter()
labels, strs, one_hot_toks = batch_converter(test_protein)
one_hot_toks
```

```python
one_hot_tokens_with_grad = one_hot_toks.requires_grad_()
out = esm2(one_hot_tokens_with_grad)["logits"].mean()
torch.autograd.grad(out, one_hot_tokens_with_grad)
```
